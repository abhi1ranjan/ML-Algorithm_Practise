## K Means Cluster Creation

Now it is time to create the Cluster labels!


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
```

Creating an instance of a K Means model with 2 clusters.

```python
kmeans = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label.**

```python
# Drop non-numeric columns (e.g., college names) and encode 'Private'
CollegeData_numeric = CollegeData.drop(['CollegeName'], axis=1)  # replace 'CollegeName' with your actual column name if different
CollegeData_numeric['Private'] = LabelEncoder().fit_transform(CollegeData['Private'])

# Define and fit the KMeans model
kmeans = KMeans(n_clusters=2)
kmeans.fit(CollegeData_numeric)
```

Getting the cluster center vectors.

```python
kmeans.cluster_centers_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

*** Creating a new column for CollegeData called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
CollegeData['Cluster'] = CollegeData['Private'].apply(converter)
```

```python
CollegeData.head()
```

** Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
print(classification_report(CollegeData['Cluster'], kmeans.labels_))
print(confusion_matrix(CollegeData['Cluster'], kmeans.labels_))
```

Not so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups!
